```python
from urllib.request import urlopen
```

```python
from bs4 import BeautifulSoup
```

```python
html = urlopen('https://www.redrickshaw.com/collections/treats-drinks')
```

```python
html.status
```

```python
bs = BeautifulSoup(html,'html.parser')
```

```python
nameList = bs.findAll('span',{'class':'rr-custom-product-item-title card-information__text h5'})
```

```python
for i in nameList:
    
    print(i.get_text())
```

```python
price = bs.findAll('span',{'class':'special-price'})
```

```python
for j in price:
    print(j.get_text())
    
```

```python
header = bs.findAll(['h1','h2','h3','h4','h5','h6'])
```

```python
for a in header:
    print(a.get_text())
```

```python
title = bs.find('title')
```

```python
print(title.text)
```

```python
paragraph = bs.find('p',{'class': 'rr-custom-collection-product-count collection-product-count light'})
```

```python
print(paragraph.text)
```

```python
link = bs.find('a',{'href':'#MainContent'})
```

```python
print(link.text)
```

```python
links = bs.findAll('div',{'id':'shopify-section-template--15204334731451__product-grid','class':'shopify-section spaced-section collection-grid-section'})
```

```python
links
```

```python
for linked in links:
    print(linked.get_text())
```

```python
print(div.get_text())
```
