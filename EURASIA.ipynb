```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import csv
```

```python
html = urlopen('https://eurasiaonline.ie/')
html = html.read()
bs = BeautifulSoup(html,'html.parser')
```

```python
for nextPage in bs.findAll('li',{'class' : 'navPages-item'}):
    print(nextPage.a.attrs['href'])
    
    
    
```

```python
sublinks = []
for nextPage in bs.findAll('li',{'class' : 'navPage-subMenu-item'}):
    nextPage = nextPage.a.attrs['href']
    sublinks.append(nextPage)
    
    
    
```

```python
print(sublinks)
```

```python
for i in sublinks:
    html = urlopen(i)
    html = html.read()
   
    bs = BeautifulSoup(html,'html.parser')
    prdLink = bs.findAll('figure', {'class' : 'card-figure'})
    for j in prdLink:
        with open('links.csv', 'a', newline='') as f:
            data_handler = csv.writer(f, delimiter=",")
            data_handler.writerow([j.a.attrs['href']])
    
    try:    
        nextPage = bs.find('li',{'class' : "pagination-item pagination-item--next"})
        nextPage = nextPage.a.attrs['href']
    except:
        continue
    
    while True:
        html = urlopen(nextPage)
        html = html.read()
        bs = BeautifulSoup(html,'html.parser')
        prdLink = bs.findAll('figure', {'class' : 'card-figure'})
        
        for j in prdLink:
            with open('links.csv', 'a', newline='') as f:
                data_handler = csv.writer(f, delimiter=",")
                data_handler.writerow([j.a.attrs['href']])
                
        try:    
            nextPage = bs.find('li',{'class' : "pagination-item pagination-item--next"})
            nextPage = nextPage.a.attrs['href']
        except:
            break
        
```

```python
for i in sublinks:
    html = urlopen(i)
    html = html.read()
   
    bs = BeautifulSoup(html,'html.parser')
    products = bs.findAll('h4',{'class' : 'card-title'})
    prices = bs.findAll('div',{'class':'card-text'})
    
    
    for i in range(0, len(products)):
        name = products[i].text
        name = name.strip()
        print(name)

        price = prices[i].text
        price = price.strip()
        print(price)

        with open('product.csv','a', newline='') as f:
            data_handler = csv.writer(f, delimiter=",")
            data_handler.writerow([name,price])
        
    try:
        nextPage = bs.find('li',{'class' : "pagination-item pagination-item--next"})
        nextPage = nextPage.a.attrs['href']
        
    except:
        continue
    
    while True:
        html = urlopen(nextPage)
        html = html.read()
        bs = BeautifulSoup(html,'html.parser')

        products = bs.findAll('h4',{'class' : 'card-title'})
        prices = bs.findAll('div',{'class':'card-text'})

        for i in range(0, len(products)):
            name = products[i].text
            name = name.strip()
            print(name)
            
            price = prices[i].text
            price = price.strip()
            print(price)
            
            with open('product.csv','a', newline='') as f:
                data_handler = csv.writer(f, delimiter=",")
                data_handler.writerow([name,price])
            

        try:    
            nextPage = bs.find('li',{'class' : "pagination-item pagination-item--next"})
            nextPage = nextPage.a.attrs['href']
        except:
            break

        
      
        
   
```
