```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
html = urlopen('https://eurasiaonline.ie/')
html = html.read()
bs = BeautifulSoup(html,'html.parser')
```

```python
for nextPage in bs.findAll('li',{'class' : 'navPages-item'}):
    print(nextPage.a.attrs['href'])
    
    
    
```

```python
sublinks = []
for nextPage in bs.findAll('li',{'class' : 'navPage-subMenu-item'}):
    nextPage = nextPage.a.attrs['href']
    sublinks.append(nextPage)
    
    
    
```

```python
print(sublinks)
```

```python

for i in sublinks:
    html = urlopen(i)
    html = html.read()
   
    bs = BeautifulSoup(html,'html.parser')
    products = bs.findAll('h4',{'class' : 'card-title'})
    prices = bs.findAll('div',{'class':'card-text'})
    
    for i in range(0, len(products)):
        name = products[i].text
        price = prices[i].text
        print(name.strip())
        print(price.strip())
        
      
            
           
        
    
        
            
      
   
```
