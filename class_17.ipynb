# reading

```python
import csv
```

```python
with open("name.csv") as f:
    file_content = csv.reader(f)
    for i in file_content:
        print(i)
```

```python
file_content
```

```python
with open("name.csv") as f:
    content = f.readable()
    
    
```

```python
print(content) #true is showing that file is readable and it contains data in it.
```

```python
with open("name.csv") as f:
    file_content = csv.reader(f)
    contents = []
    for i in file_content:
        contents.append(i)
       
```

```python
print(contents) #list of list
```

```python
with open("name.csv") as f:
    file_content = csv.reader(f)
    contents = []
    for i in file_content:
        for j in i:
            contents.append(j)
       
```

```python
print(contents) #single list using nested loop
```

# writing

```python
with open("hunain.csv","w",newline = "") as f:
    data_write = csv.writer(f,delimiter = ',')
    data_write.writerow(['Year', 'Event', 'Winner'])
    data_write.writerow(['1995', 'Best-Kept Lawn', 'None'])
    
    
```

# appending

```python
with open("hunain.csv","a",newline = "") as f:
    data_write = csv.writer(f,delimiter = ',')
    data_write.writerow(['1999', 'Gobstones', 'Welch National'])
    
```
